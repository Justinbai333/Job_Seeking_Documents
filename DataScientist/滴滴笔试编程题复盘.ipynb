# 滴滴笔试复盘

编程题：

题目，给定几个连续的柱子高度，如1, 4, 2, 3, 5, 3, 3
现在往这些柱子中间注水，问可以注多少水不溢出来

如：1,4,2,3,5,3,3
可以注入的水量分别为1 (1) 4 (4) 2 (4) 3 (4) 5 (3) 3 (3) 3
输出：19(括号内的数字和)

```python
#先把整个表化成一个int list
mylist = []
for i in input().split(','):
    mylist.append(int(i))

#设定两个初始值，tallest为local的最高，初始为第一根柱子的高度，因为下面的循环从第二个开始循
sum = 0
tallest = int(mylist[0])


for i in range (1,len(mylist)):
    
    #判断和前面的local tallest的关系
    #如果新柱子比前面的local tallest高，注水量为之前local tallest的高度，更新local tallest为自己
    if mylist[i] >= tallest:
        sum = sum + tallest
        tallest = mylist[i]

    else:
        #如果新柱子比local tallest矮，分两种情况，判断后面是否还有更高 / 等高的柱子
        taller_exist = False
        for j in range(i, len(mylist)):
            if mylist[j] >= tallest:
                taller_exist = True
        
        #如果之后有，则注水量为local tallest，且不更新local tallest
        if taller_exist:
            sum = sum + tallest
        
        #如果之后没有，则注水量为新柱子的高度，且更新local tallest
        else:
            sum = sum + mylist[i]
            tallest = mylist[i]

print(sum)
```
